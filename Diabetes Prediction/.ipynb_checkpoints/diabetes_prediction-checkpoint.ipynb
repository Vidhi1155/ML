```python
class KNNClassifier:
    def __init__(self, k=5):
        self.k = k

    def fit(self, x, y):
        self.X = x            
        self.y = y            

    def distance_formula(self, p1, p2):
        return np.sqrt(np.sum((p1 - p2)**2))

    def predict_1(self, t):
        distances = []          
        for i in range(len(self.X)):
            distances.append((self.distance_formula(t, self.X[i]), self.y[i]))
            
        distances = sorted(distances, key = lambda x : x[0])[: self.k]       
        klabels = []                   
        for i in range(self.k):
            klabels.append(distances[i][1])
            
        labels, counts = np.unique(klabels, return_counts=True)            
        predLabel = labels[np.argmax(counts)]                             
        prob = np.max(counts)/self.k     
        
        return predLabel

    def predict(self, t):
        predictions = []
        for test in t:
            predLabel = self.predict_1(test)
            predictions.append(predLabel)
                
        return np.array(predictions)

    def accuracy(self, y_pred, y_test):
        return np.sum(y_pred == y_test)/y_test.shape[0]
```

```python
x = pd.read_csv('diabetes.csv', usecols=[0,1,2,3,4,5,6,7])
y = pd.read_csv('diabetes.csv', usecols=[8])
x_data, y_data = x.values, y.values.reshape(-1)

std = StandardScaler()     
x_data = std.fit_transform((x_data))
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, train_size = 0.8, random_state= 90)
```

```python
clf = KNNClassifier(73)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = (clf.accuracy(y_pred, y_test)) * 100
print("Accuracy:", round(acc, 3))
```

Using KNN from scikit-learn

```python
acc = []
k = []
for i in range(1, 301):
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    acc.append(accuracy_score(y_test, y_pred))
    k.append(i)

plt.plot(k, acc)         
plt.show()
```
